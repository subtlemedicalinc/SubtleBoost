```python
from glob import glob
import pydicom
import os

path = '/mnt/datasets/ben/backup/raid/SubtleGAD_Stanford/lowcon'
cases = sorted([d.split('/')[-1] for d in glob('{}/Patient*'.format(path))])
```

```python
for cnum in cases:
    series_dirs = sorted([d for d in glob('{}/{}/*'.format(path, cnum)) if os.path.isdir(d)])
    print('\n=====', cnum, '=====')
    for sdir in series_dirs:
        fpath_dcms = [f for f in glob('{}/*.dcm'.format(sdir))]
        if len(fpath_dcms) == 0:
            continue
        hdr = pydicom.dcmread(fpath_dcms[0])
        try:
            print(hdr.SeriesDescription)
        except Exception as e:
            print('NOT DICOM')
```
