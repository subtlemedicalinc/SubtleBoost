```python
def preprocess_data(data, con_idxs=[0, 2, 3, 4]):
    data = data / data.mean()
    data = data[con_idxs, ...]
    data = np.rot90(data, axes=(2, 3))
    
    s_idx = int(data.shape[1] * 0.1)
    data = data[:, s_idx:-s_idx, 48:-48, 40:-24]
    data = np.clip(data, 0, data.max())
    data = data.astype(np.float32)
    return data

def pp_data_fb(data, con_idxs=[0, 2, 3, 4]):
    data = data[con_idxs, ...]
    for idx in np.arange(data.shape[0]):
        data[idx] = data[idx] / data[idx].mean()
        data[idx] = np.clip(data[idx], 0, data[idx].max())
    data = np.rot90(data, axes=(2, 3))
    data = np.array([zoom(data[c], zoom=(1, 0.875, 0.875)) for c in np.arange(data.shape[0])])
    
    data = data[:, :, 32:-32, 24:-8]
    data = np.clip(data, 0, data.max())
    data = data.astype(np.float32)
    return data


def visualize_data(patient_id, sl_idx, 
                   src_path='/home/srivathsa/projects/studies/gad/stanford/preprocess/data_mmt/zero_gad/train'):
    fpath_npy = os.path.join(src_path, patient_id, '{:03d}.npy'.format(sl_idx))
    img = np.load(fpath_npy)
    img_stack = [d for d in img]
    img_disp = np.hstack(img_stack)
    plt.imshow(img_disp)
```
