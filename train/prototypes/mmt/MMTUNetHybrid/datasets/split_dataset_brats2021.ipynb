```python
#fix datasplit
import pickle
import random
import os
with open('data_split_brats2021.pickle', 'rb') as handle:
    data = pickle.load(handle)
cases = data['test_cases'] + data['val_cases']
random.shuffle(cases)
```

```python
n_test = len(cases)//2
test_cases = cases[:n_test]
val_cases = cases[n_test:]
data_split = {'train_cases': data['train_cases'], 'val_cases': val_cases, "test_cases": test_cases}
with open('data_split_brats2021.pickle', 'wb') as handle:
    pickle.dump(data_split, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
